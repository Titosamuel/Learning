# SOBREAJUSTE Y SUBAJUSTE

**SUBAJUSTE** Se refiere a un modelo que no puede modelar los datos de entrenamiento o no generalizar los nuevos datos.

Esto ocurre cuando el modelo de Machine Learning es muy simple.

*Su aparición simplemente significa que nuestro modelo o el algoritmo, no se ajusta a los datos lo suficientemente bien.*

**Sucede cuando:**

- Tenemos menos datos para construir un modelo preciso.
- Cuando intentamos construir un modelo lineal con datos no lineales.

**SOBREAJUSTE** Se refiere a un modelo que modela los datos de entrenamiento demasiado bien.

Esto ocurre cuando el modelo de Machine Learning aprende el detalle incluyendo el ruido en los datos de entrenamiento en la medida que tiene un impacto negativo en el modelo de los datos de entrenamiento.

Esto significa que el ruido o las fluctuaciones aleatorias en los datos de entrenamiento, son recogidos y aprendidos por el modelo.

*Es más probable con modelos no paramétricos y no lineales*

**COMO EVITAR EL SOBREAJUSTE Y EL SUBAJUSTE**

**Usar más información para entrenamiento:** El uso de un gran conjunto de datos de capacitación, generalmente ayuda al modelo de Machine Learning a alegir la señal de manera eficiente, sin embargo, esta técnica puede no funcionar todas las veces. Si se agregan muchos datos ruidosos, y los datos relevantes son escasos, no ayudará al modelo a predecir correctamente.

**Técnica de validación cruzada:** Si tienes los datos, usar un conjunto de datos de validación también es una práctica excelente. Una forma estándar de encontrar un error de predicción fuera de muestra es usar una validación cruzada de 5 veces.

**Detección temprana:** Cuando entrenas un modelo, observarás que hasta cierto número de pruebas, el rendimiento del modelo mejora. Después el modelo tendrá un rendimiento bajo con los datos de prueba. Se debe detener las iteraciones del modelo, antes que exista un sobreajuste excesivo.

**Regularización:** La regularización se hace para simplificar el modelo de ML y consiste en muchos métodos. La técnica de regularización utilizada para un modelo de ML, depende del tipo de modelo. Por ejemplo, si el modelo es un árbol de decisión, la regularización sería podar el árbol. Si el modelo es una regresión, se puede agregar una penalización al valor de costo de regularización.
