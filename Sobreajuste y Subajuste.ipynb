# SOBREAJUSTE Y SUBAJUSTE

**SUBAJUSTE** Se refiere a un modelo que no puede modelar los datos de entrenamiento o no generalizar los nuevos datos.

Esto ocurre cuando el modelo de Machine Learning es muy simple.

*Su aparicin simplemente significa que nuestro modelo o el algoritmo, no se ajusta a los datos lo suficientemente bien.*

**Sucede cuando:**

- Tenemos menos datos para construir un modelo preciso.
- Cuando intentamos construir un modelo lineal con datos no lineales.

**SOBREAJUSTE** Se refiere a un modelo que modela los datos de entrenamiento demasiado bien.

Esto ocurre cuando el modelo de Machine Learning aprende el detalle incluyendo el ruido en los datos de entrenamiento en la medida que tiene un impacto negativo en el modelo de los datos de entrenamiento.

Esto significa que el ruido o las fluctuaciones aleatorias en los datos de entrenamiento, son recogidos y aprendidos por el modelo.

*Es ms probable con modelos no paramtricos y no lineales*
