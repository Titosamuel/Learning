# Ejercicio Boston Houses

```python
#Se importa la librería a utilizar
from sklearn import datasets, linear_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importamos los datos de la misma librería de scikit-learn
boston = datasets.load_boston()
print(boston)
print()
```

```python
#Verificar la información contenida en el dataset
print('Información en el dataset:')
print(boston.keys()) #target, son las etiquetas o respuestas
print()
```

```python
#verifico las características del dataset
print('Características del dataset:')
print(boston.DESCR)
```

```python
#Verifico la cantidad de datos que hay en los dataset
print('Cantidad de datos:')
print(boston.data.shape)
print()
```

```python
#Verifico la información de las columnas
print('Información de las columnas:')
print(boston.feature_names)
print()
```

```python
#Seleccionamos las columnas 5,6 y 7 del dataset
X_multiple = boston.data[:, 5:8]
print(X_multiple)
```

```python
#Definir los datos correspondientes a las etiquetas (target)
Y_multiple = boston.target
```

<hr>

**IMPLEMENTACIÓN DE REGRESIÓN LINEAL MULTIPLE**

```python
from sklearn.model_selection import train_test_split

#separo los datos de "train" en entrenamiento y "prueba"
#para probar los algoritmos
X_train,X_test,Y_train,Y_test = train_test_split(X_multiple,Y_multiple, test_size=0.2)
```

```python
#Defino el algoritmo a utilizar
lr_multiple = linear_model.LinearRegression()
```

```python
#entreno el modelo
lr_multiple.fit(X_train, Y_train)
```

```python
#Realizo la predicción
Y_pred_multiple = lr_multiple.predict(X_test)
```

```python
print('DATOS DEL MODELO REGRESIÓN LINEAL MÚLTIPLE')
print()

print('Valor de las pendientes o coeficientes "a":')
print(lr_multiple.coef_)
```

```python
#Calculamos la intersección o el valor de "b"
print('Valor de las intersecciones o coeficientes "b":')
print(lr_multiple.intercept_)
```

```python
#Ahora calculamos la precisión del algoritmo
print('Precisión del algoritmo: datosTrain')
print(lr_multiple.score(X_train, Y_train))
print()
print('Precisión del algoritmo: datosTest')
print(lr_multiple.score(X_test, Y_test))
```

**Este modelo es muy parecido a la regresión lineal** la úunica diferencia es que se agregan más variables en X
