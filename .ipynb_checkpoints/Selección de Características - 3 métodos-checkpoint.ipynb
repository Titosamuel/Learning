# Selección de características

Acá se mostrarán 3 métodos:
- Métodos de Filtros
- Métodos Integrados
- Métodos de Envoltura

Porque se debe implementar la selección de características en ML?


 **QUE ES LA SELECCIÓN DE CARACTERÍSTICAS?** Es el proceso de seleccionar las más importantes y/o relevantes característica de un conjunto de datos, con el objetivo de mejorar el rendimiento de predicciön de los predictos, proporcionar predictores más rápidos y más rentables. Y además, proporcionar una mejor comprensión del proceso subyacente que generó los datos.

<hr>

**Métodos de Filtros:**
Se utilizan generalmente como un paso de preprocesamiento de datos. La selección de características es independiente de cualquier algoritmo de ML.

Las características se clasifican según los puntajes estadísticos que tienden a determinar la correlación de las características con la variable de resultado.

La siguiente tabla puede ser utulizada para definir los coeficientes de correlación, para los diferentes tipos de datos.


**Correlación de Pearson:** Se usa conmo una medida para cuantificar la dependencia lineal entre dos variables continuas X e Y. Su valor varía de -1 a 1.

**LDA:** El análisis discriminante Lineal, se usa para encontrar una combinación lineal de características. Que caracteriza o sepra dos o más clases, o niveles de una variable categórica.

**Anova (Análisis de la varianza):** Es similar a LDA, excepto por el hecho de que opera mediante una o más fucniones independientes categóricas, y una función dependiente contínua. Proporciona una medida si varios grupos son iguales o no.

**Chi-cuadrado:** Prueba estadística que se aplica a los grupos de características categóricas para evaluar la probabilidad de correlación o asociación entre ellos, utilizando su distribución de frecuencia.

Estos métodos no eliminan la multicolinealidad.

<hr>

**Método de envoltura:** Busca una característica que sea más adecuada para el algoritmo y tiene como objetivo, mejorar el rendimiento. Son métodos computacionalmente muy caros.


**Selección hacia delante:** Es un método iterativo en el que comenzamos sin tener ninguna característica en el modelo. En cada iteración, seguimos agregando la función que proporciona mejoras a nuestro modelo, hasta que la adición de una nueva variable no mejore el rendimiento del modelo.

**Eliminación hacia atrás:** Comenzamos con todas la características menos significativas en cada iteración, lo que mejora el rendimiento del modelo. Repetimos esto hasta que no se observe ninguna mejora en la eliminación de características.

**Eliminación de características recursivas:** Es un algoritmo de optimización que busca encontrar el subconjunto de funciones con mejor rendimeinto. Crea repetidamente modelos y deja de lado la mejor o la peor característica de rendimiento en cada iteración.

<hr>

**Métodos Integrados:** Combina las cualidades de los métodos de filtro y envoltura. Se implementa mediante algoritmos que tienen sus propios métodos de selección de características incorporados.

Algunos de los ejemplos más populares de estos métodos son la regresión LASSO y RIDGE.
