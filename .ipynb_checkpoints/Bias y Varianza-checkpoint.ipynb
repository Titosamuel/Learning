**TIPOS DE ERRORES**

1. Errores de bias(Sezgo)
2. Errores de varianza
3. Errores irreducibles

<hr>

**3. ERRORES IRREDUCIBLES**
    - No se puede reducir
    - Se les conoce como ruido
    - Provienen de muchos factores
        
        
        -Variables desconocidas (in/out)
        -Características incompletas
        -Problema mal enmarcado

*No importa cuán bueno hagamos nuestro modelo, nuestros datos tendrán cierta cantidad de ruido o un error irreductible que no se puede eliminar*

**1. ERRORES DE BIAS(SEZGO)**

Es la diferencia entre la predicción esperada de nuestro modelo y los valores verdaderos.

*Los algoritmos paramétricos tienen un alto bias que los hace rápidos de aprender y más fácil de entender, pero generalmente menos flexibles.*

*Tienen un menor rendimiento predictivo en problemas complejos*



**Bajo BIAS:** Sugiere menos suposiciones sobre la forma de la función objetivo

Algoritmos
- Árboles de decisión
- K-Vecinos más cercanos
- Máquinas de vectores de soporte

**Alto BIAS:** Sugiere más suposiciones sobre la forma de la función objetivo

Algoritmos
- Regresión lineal
- Análisis discriminante lineal
- Regresión logística

**2. ERRORES DE VARIANZA**

Se refiere a la cantidad que la estimación de la función objetivo cambiará si se utiliza diferentes datos de entrenamiento.

Idealmente no debería cambiar demasiado de un conjunto de datos de entrenamiento a otro

**Varianza BAJA:** Sugiere pequeños cambios en la estimación de la función objetivo, con cambios en el conjunto de datos de capacitación

Algoritmos
- Regresión lineal
- Análisis discriminante lineal
- Regresión logística


**Varianza ALTA:** Sugiere grandes cambios en la estimación de la función objetivo con cambios en el conjunto de datos de capacitación

Algoritmos
- Árboles de decisión
- K-Vecinos más cercanos
- Máquinas de vectores de soporte

**Aumentar el BIAS, disminuirá la VARIANZA**

**Aumenta la VARIANZA, disminuirá el BIAS**

En general, lo que realmente importa es el error general, no la composición específica.

*Para construir un buen modelo, necesitamos encontrar un buen equilibrio entre el BIAS y la VARIANZA, de manera que minimice el ERROR TOTAL.*

Comprender el BIAS y la VARIANZA es fundamental para comprender el comportamiento de los modelos de predicción
