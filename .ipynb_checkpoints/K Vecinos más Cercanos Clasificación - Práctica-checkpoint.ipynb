```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
#Obtenemos el dataset de CANCER de la librer√≠a de SCIKITLEARN
dataset = datasets.load_breast_cancer()
print(dataset.keys())
```

```python
#Seleccionamos todas las columnas
X = dataset.data

#Defino los datos correspondientes a las etiquetas
Y = dataset.target
```

**IMPLEMENTACI√ìN DE K VECINOS M√ÅS CERCANOS**

```python
from sklearn.model_selection import train_test_split

#Separo los datos de "train" en entrenamiento y prueba
#para poder probar los algoritmos

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
#Defino el algoritmo a utilizar
from sklearn.neighbors import KNeighborsClassifier
```

```python
algoritmo = KNeighborsClassifier(n_neighbors= 5, metric='minkowski', p=2)
#estas m√©tricas viene por defecto de esta forma
#si quisieramos, se podr√≠a dejar el par√©ntesis vac√≠o
#pero se muestran las 3 m√©tricas a modod e ense√±anza
```

```python
#entreno el modelo
algoritmo.fit(X_train,Y_train)
```

```python
#Realizo una predicci√≥n
Y_pred = algoritmo.predict(X_test)
```

```python
#Para validar que el modelo funciona, se verifica la
#Matr√≠z de confusi√≥n (ver en video de curso)

from sklearn.metrics import confusion_matrix

matriz = confusion_matrix(Y_test, Y_pred)
print('Matriz de COnfusi√≥n:')
print(matriz)
```

**La diagonal Principal** indica la cantidad de datos predichos correctamente, y **La diagonal secundaria** son los errores. En este caso es mayor la cantidad de datos correctos que incorrectos. La suma de la diagonal principal es 109 e incorrectos 5.

```python
#Validar la precisi√≥n del modelo
from sklearn.metrics import precision_score

precision = precision_score(Y_test, Y_pred)
print('Precisi√≥n del modelo:')
print(precision)
```
