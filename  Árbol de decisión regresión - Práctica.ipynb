# Práctica

```python
#importar todos los paquetes que se necesitan
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from IPython.display import display, HTML
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
#se importa el dataset especīfico de boston
boston = datasets.load_boston()
print(boston.feature_names)
```

**PREPARAR LA DATA PARBOLES DE DECISIÓN REGRESIÓN**

```python
#seleccionamos solamente la columna 6 del dataset
X_adr = boston.data[:, np.newaxis, 5]

#Defino los datos correspondientes a las etiquetas
#(Y o variables dependientes)
Y_adr = boston.target
```

```python
#Graficar los datos correspondientes
plt.scatter(X_adr,Y_adr)
plt.xlabel('Variable Independiente(Prom.Núm.Piezas)')
plt.ylabel('Variable Dependiente(Target)')
plt.show()
```

**IMPLEMENTACIÓN DE ÁRBOLES DE DECICIÓN REGRESIÓN**

```python
from sklearn.model_selection import train_test_split

#Separo los datos de "train" en entrenamiento
#y prueba para probar los algoritmos
X_train, X_test, Y_train, Y_test = train_test_split(X_adr, Y_adr, test_size=0.2)

from sklearn.tree import DecisionTreeRegressor

#defino el algoritmo a utilizar
adr = DecisionTreeRegressor(max_depth=10)
```

```python
#Entreno el modelo
adr.fit(X_train, Y_train)
```

```python
#Realizo una predicción
Y_pred = adr.predict(X_test)
```

```python
#comparo el test con la predicción
print(Y_test)
```

```python
print(Y_pred)
```

```python
#Graficamos los datos de prueba junto con la predicción
X_grid = np.arange(min(X_test), max(X_test), 0.1)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X_test, Y_test)
plt.plot(X_grid, adr.predict(X_grid), color='red', linewidth=3)
plt.show()
```

```python
#Evaluamos la precisión del modelo
print('DATOS DEL MODELO ÁRBOLES DE DECISIÓN REGRESIÓN')
print()
print('Precisión del modelo:')
print(adr.score(X_train,Y_train))
print(adr.score(X_test, Y_test))
```
