# Regresión Lineal

Técnica paramétrica utilizada para predecir variables contínuas, dependientes, dado un conjunto de variables independientes.

De naturaleza paramétrica, porque hace ciertas suposiciones basadas en el conjunto de datos.

*Pendiente se le conoce como coeficiente, especie de magnitud de cambio*

El objetivo de la regresión lineal es minimizar la distancia de los datos con la linea.


**TÉCNICA PARA REDUCIR EL ERROR**


**Mínimos Cuadrados:** Intenta reducir la suma de los errores, al cuadrado.

Utiliza un error cuadrado que tiene buenas propiedades matemáticas.

Es fácil de analizar y computacionalmente más rápido.

La interpretación es mucho más fácil que otras técnicas.
